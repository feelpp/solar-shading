# Data format and visualisation

## Data format

Solar masks are currently stored in CSV files as dense matrices of dimensions $N_{Az} \times N_{Alt}$, where $N_{Az}$ and $N_{Alt}$ correspond to the number of intervals discretising the azimuth and altitude angular coordinates.
The entries of the shading mask matrix correspond to the percentage of direct solar radiation that is blocked by the environment surrounding the object. Their values belong to the interval $[0,1]$, where the extrema correspond, respectively, to absence of occlusion and to complete occlusion.
## Visualisation
